```python
import pandas as pd
```

```python
carseats = pd.read_csv('https://raw.githubusercontent.com/JWarmenhoven/ISLR-python/master/Notebooks/Data/Carseats.csv')
```

```python
%matplotlib inline
import pandas as pd
import numpy as np
import scipy.stats as stats
import statsmodels.api as sm
import statsmodels.formula.api as smf
import matplotlib.pyplot as plt
```

```python
plt.figure()
pd.plotting.scatter_matrix(carseats, alpha = 0.4)
plt.show()
```

```python
my_lm = smf.ols("Sales ~ Price + Urban + US", carseats).fit()
print(my_lm.summary())
```

```python
my_lm = smf.ols("Sales ~ Price + US", carseats).fit()
print(my_lm.summary())
```

```python
import statsmodels.stats as smstats
smstats.outliers_influence.OLSInfluence(my_lm)
```
